```python
# Etapa 1: Preparar o Terreno (Inputs e Parâmetros)
# conectar a api e colocar parametros
import requests
import pandas as pd

lat = float(input('Informe a latitude desejada: '))
lon = float(input('Informe a longitude desejada: '))
quantos_dias = 7

url = 'https://api.open-meteo.com/v1/forecast'

parametros = {
    'latitude' : lat,
    'longitude': lon,
    'daily': ["temperature_2m_max", 'precipitation_sum'],
    'timezone' : 'auto',
    "past_days" : quantos_dias,
    'forecast_days': 0
}
# Etapa 2: Coleta de Dados e Verificação
# usar o get
resposta = requests.get(url, params=parametros)
# if do status_code
if resposta.status_code == 200:

  # salvar no json
  dados_completos = resposta.json()
  dados_diarios = dados_completos['daily']

  # criar o dataframe como pandas
  df_dados_clima = pd.DataFrame(dados_diarios)

  # EXTRA - mostrar no padrao dos EUA
  # (temperatura * 1.8) + 32
  df_dados_clima['temperature_2m_max_fahrenheit'] = ((df_dados_clima['temperature_2m_max'] * 1.8) + 32).round(2)

  # se quiser display(dataframe)
  display(df_dados_clima)


  # Etapa 3: Salva os dados (o arquivo CSV)
  df_dados_clima.to_csv('relatorio_prof_luiz.csv', index=False)
  # salva o arquivo
else:
  print(f"Erro na requisição: {resposta.status_code}")
```
